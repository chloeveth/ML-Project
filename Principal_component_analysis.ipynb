```python
%matplotlib inline
import mglearn
mglearn.plots.plot_pca_illustration()
```

Let's think through was this example means. Focusing just on the original data in the top left,
remember that the principal components themselves are vectors where the first principal component is the most
informative direction, the second the next most informative, etc. Each next principal component is orthogonal to
all the preivous. 

The upper right plot shows the data plotted on the principal components rather than the original axes of the features. This effectively rotates and centers the data.

Now, that doesn't reduce the dimensionaly---we had two features to begin with, and we found two principal components. To reduce dimensionality, project the data onto the first principal component, as seen in the lower right and left,
with the lower right specifically showing that projection back in the original feature space. 
Notice that although this means we dropped a principal component, we do not drop either of the original features. The data points' (now one-dimensional) positions are calculated from both original features.

## 2. Using PCA to look at the data

One of the first steps in any machine learning project is, *look at the data*. 
Mchine learning tools aren't oracles into which you pour data, press a button,
and get an automatic answer.
They are flexible tools that require some understanding of the data in order to use them
appropriately.
Do you remember this visualization of the iris data set?


```python
import pandas as pd
from sklearn.datasets import load_iris
iris_dataset = load_iris()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

# create dataframe from data in X_train
# label the columns using the strings in iris_dataset.feature_names
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
# create a scatter matrix from the dataframe, color by y_train
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15),
                           marker='o', hist_kwds={'bins': 20}, s=60,
                           alpha=.8, cmap=mglearn.cm3)
```

The plots in this matrix show pairs of features plotted together to help make hypotheses about
correlations and what features of combinations may be informative for classification.
This is feasible if we narrow our view to just four features, as in this case.
This sort of thing would be harder to scale to, for example, all 30 features of the breast cancer data set.

We can visualize the data using other plots, such as these feature-histograms for the brest cancer data set.
The two colors indicate the classes, malignant and benign.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

fig, axes = plt.subplots(15, 2, figsize=(10, 20))
malignant = cancer.data[cancer.target == 0]
benign = cancer.data[cancer.target == 1]

ax = axes.ravel()

for i in range(30):
    _, bins = np.histogram(cancer.data[:, i], bins=50)
    ax[i].hist(malignant[:, i], bins=bins, color=mglearn.cm3(0), alpha=.5)
    ax[i].hist(benign[:, i], bins=bins, color=mglearn.cm3(2), alpha=.5)
    ax[i].set_title(cancer.feature_names[i])
    ax[i].set_yticks(())
ax[0].set_xlabel("Feature magnitude")
ax[0].set_ylabel("Frequency")
ax[0].legend(["malignant", "benign"], loc="best")
fig.tight_layout()
```

This can give us some sense of the informativeness of individual readings (mean concave points is informative, symmetry error is not), but it doesn't help us identify combinations of features that might be useful for diagnosis. 
This is one place were PCA can help us. 

First, normalize the various features so they all have variance 1. 
This is what `scikitlearn`'s 
[`StandardScalar`](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html) does.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler

cancer = load_breast_cancer()

scaler = StandardScaler()
scaler.fit(cancer.data)
X_scaled = scaler.transform(cancer.data)
```

`scikitlearn`'s `PCA` class is used for the analysis. Indicate the number of principal components when
instantiating the class.
Then the `fit` method finds the components.
Let's take a look at the components themselves, as well as how informative they are,
indicated by the `explained_variance_` attribute.

```python
from sklearn.decomposition import PCA
# keep the first two principal components of the data
pca = PCA(n_components=2)
# fit PCA model to beast cancer data
pca.fit(X_scaled)

print(pca.components_)
print(pca.explained_variance_)
```

"Explained variance" means (as its name suggests), how much of the variance in the data is "explained" by that component, that is, how much the variation occurs along that direction. This "explained variance" is equal to the eigenvalue corresponding to the eigenvector that is this component. (See [the sklearn PCA documentation](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html).

Now we use that object to transform the data.

```python
# transform data onto the first two principal components
X_pca = pca.transform(X_scaled)
print("Original shape: {}".format(str(X_scaled.shape)))
print("Reduced shape: {}".format(str(X_pca.shape)))
```

Let's see what the data looks like when plotted on just those two principal components.

```python
# plot first vs. second principal component, colored by class
plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_pca[:, 0], X_pca[:, 1], cancer.target)
plt.legend(cancer.target_names, loc="best")
plt.gca().set_aspect("equal")
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```

The data becomes almost completely separable. Remember that PCA doesn't itself do classification---but it seems
that the transformed data is very ammenable to such classification techniques as KNN or SVM.

One might reasonably wonder if we can learn something about the data from the principal components themselves.
Here is a "heat map" for the two principal components. See if you can conclude anything from this visualization:

```python
plt.matshow(pca.components_, cmap='viridis')
plt.yticks([0, 1], ["First component", "Second component"])
plt.colorbar()
plt.xticks(range(len(cancer.feature_names)),
           cancer.feature_names, rotation=60, ha='left')
plt.xlabel("Feature")
plt.ylabel("Principal components")
```

Here is what M&uuml;ller and Guido say:

> You can see that in the first component, all features have the same sign
(it's positive, but it doesn't matter which direction the arrows point in).
That means that there is a general correlation between all features.
As one measurement is high, the others are likely to be hight as well.
The second component has mixed signs, and both of the components involve all of the 30 features.
This mixing of all features is what makes explaining the
axes tricky. (Pg 149.)

Here is a selection of the names together with their frequency:

```python
# count how often each target appears
counts = np.bincount(people.target)
# print counts next to target names:
for i, (count, name) in enumerate(zip(counts, people.target_names)):
    print("{0:25} {1:3}".format(name, count), end='   ')
    if (i + 1) % 3 == 0:
        print()
```

(You'll notice that it's very male-heavy, and the females tend to be actresses and atheletes, whereas the males tend to be politicians. Maybe that's an ethics discussion we need to have. But it's the dataset we have. Also, the frequencies follow Zipf's law.)

Let's restrict this so we use no more than 50 images of any person.

```python
mask = np.zeros(people.target.shape, dtype=np.bool)
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1
    
X_people = people.data[mask]
y_people = people.target[mask]

# scale the grey-scale values to be between 0 and 1
# instead of 0 and 255 for better numeric stability:
X_people = X_people / 255.
```

One difficulty in facial recognition is the sparsity of data compared to the number of
classes. We have many people (classes), but typically few images per class.
Consider the performance of a classifier that looks only at the single closest training data point for classifying a new data point: 

```python
from sklearn.neighbors import KNeighborsClassifier
# split the data in training and test set
X_train, X_test, y_train, y_test = train_test_split(
    X_people, y_people, stratify=y_people, random_state=0)
# build a KNeighborsClassifier with using one neighbor:
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Test set score of 1-nn: {:.2f}".format(knn.score(X_test, y_test)))
```

You can imagine how a pixel-by-pixel comparison would give very little information.
Let's see how PCA can help. 

First, we'll use a feature of the `PCA` class called *whitening*
which normalizes the principal components.
Consider its affect on the toy dataset from earlier:

```python
mglearn.plots.plot_pca_whitening()
```

Now with whitening, let's do a PCA on the faces dataset, and split the data into training and test set.

```python
pca = PCA(n_components=100, whiten=True, random_state=0).fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

print("X_train_pca.shape: {}".format(X_train_pca.shape))
```

How do you think this will affect the accuracy of one-nearest-neighbor?

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_pca, y_train)
print("Test set accuracy: {:.2f}".format(knn.score(X_test_pca, y_test)))
```

In interpreting that number, keep in mind the simplicity of the classifier and what we were starting from without PCA. 
Now let's see what we can learn from PCA about the images. 
Remember that images are, in this view, vectors in a space equal to the number of pixels.
That's $87 \times 65 = 5655$ in this case.
The principal components are directions within that space.
This also means that principal components themselves can be treated as images---we can
actually look at them:

```python
print("pca.components_.shape: {}".format(pca.components_.shape))
fig, axes = plt.subplots(3, 5, figsize=(15, 12),
                         subplot_kw={'xticks': (), 'yticks': ()})
for i, (component, ax) in enumerate(zip(pca.components_, axes.ravel())):
    ax.imshow(component.reshape(image_shape),
              cmap='viridis')
    ax.set_title("{}. component".format((i + 1)))
```

What is eerie about this is that these are not blurred versions of individual pictures.
These are *abstract faces*, facial aspects drawn from the set of faces as a whole.
Each image in the data set is a linear combination of these facial aspects (and the others, had we bothered to calculate all 5655 of them).
That said, these aspects are influenced more heavily by the faces of the more frequently occurring people. 
If we hadn't limited each individual to no more than 50 images, these principal components probably would end up having more resemblance to George W Bush.
(If you're doing ok on time, you may want to go back and try it.)

Can we assign meaning to any of these principal components?

M&uuml;ller and Guido say:

> While we cannot understand all aspects of these compoents, 
we can guess which aspects of the face images some of the components are capturing.
The first component seems mostly to encode the contrast between the face and the background, the second component encodes differences in lighting between the right and left half of the face, and so on. 
While this representation is slightly more semantic than the raw pixel values,
it is still quite far from how a human might perceive a face. 
As the PCA model is based on pixels, the alignment of the face (the position of eyes,
chin, and nose) and the lighting both have a strong influence on how similar two images are in their pixel representation. 
But alignment and lighting are probably
not what a human would perceive first.
When asking people to rate similarity of faces, they are more likely to 
use attributes like age, sex, facial expression, and hair style,
which are attributes that are hard to infer from the pixel intensities.
It's important to keep in mind that algorithms often interpret data 
(particularly visual data, such as images, which humans are very familiar with)
quite differently from how a human would.

It is worth nothing that features like face/background contrast or lighting are unlikely to be informative for classifying whose face the picture is (unless some people are more likely to be photographed in a dark background for some reason?). The target data does not affect finding the principal components

Consider what the faces would look like when projected onto a subset of the principal components. 
This is analogous to the "back-rotation using only first component" in the very first 
code input of this lab---scroll back to refresh your memory.

```python
reduced_images = []
for n_components in [10, 50, 100, 500]:
    # build the NMF model
    pca = PCA(n_components=n_components)
    pca.fit(X_train)
    # transform the test data (afterwards has n_components many dimensions)
    X_test_pca = pca.transform(X_test)
    # back-transform the transformed test-data
    # (afterwards it's in the original space again)
    X_test_back = pca.inverse_transform(X_test_pca)
    reduced_images.append(X_test_back)

# plot the first three images in the test set:
fix, axes = plt.subplots(3, 5, figsize=(15, 12),
                         subplot_kw={'xticks': (), 'yticks': ()})
for i, ax in enumerate(axes):
    # plot original image
    ax[0].imshow(X_test[i].reshape(image_shape),
                 vmin=0, vmax=1)
    # plot the four back-transformed images
    for a, X_test_back in zip(ax[1:], reduced_images):
        a.imshow(X_test_back[i].reshape(image_shape), vmin=0, vmax=1)

# label the top row
axes[0, 0].set_title("original image")
for ax, n_components in zip(axes[0, 1:], [10, 50, 100, 500]):
    ax.set_title("%d components" % n_components)
```

500 principal components is a lot---but it's also less than 5655, the inherent dimensionality of the data, and at 500 the faces are pretty recognizable (by humans). You can also see why the nearest-neighbor classifier isn't very good at 100 components.
